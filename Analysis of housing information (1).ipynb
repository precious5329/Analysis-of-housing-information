###[Problem 5] Researching terminology

**Kurtosis:**
Kurtosis is a statistical measure that describes the distribution of data points in a dataset. It quantifies the degree to which the tails of a distribution differ from those of a normal distribution. In other words, it measures the "tailedness" of the data distribution.

- **Leptokurtic (Positive Kurtosis):** A distribution with positive kurtosis has fatter tails and a higher peak than a normal distribution. It indicates a higher probability of extreme values.
  
- **Mesokurtic (Zero Kurtosis):** A distribution with zero kurtosis is a normal distribution, meaning that it has the same tails and peak as a normal distribution.

- **Platykurtic (Negative Kurtosis):** A distribution with negative kurtosis has thinner tails and a lower peak than a normal distribution. It suggests a lower probability of extreme values.

Kurtosis is calculated using the fourth standardized moment. Positive kurtosis values indicate heavy tails, while negative values indicate light tails.

**Skewness:**
Skewness is a measure of the asymmetry of a probability distribution. It quantifies the extent and direction of skew (departure from horizontal symmetry) in a dataset.

- **Positively Skewed (Right-skewed):** The distribution has a tail on the right side that is longer or fatter than the left side. The mass of the distribution is concentrated on the left side.

- **Negatively Skewed (Left-skewed):** The distribution has a tail on the left side that is longer or fatter than the right side. The mass of the distribution is concentrated on the right side.

- **Symmetrical (Zero Skewness):** A distribution is symmetric if the two sides of the histogram are mirror images of each other. It has zero skewness.

Skewness is calculated using the third standardized moment. It provides insights into the shape and direction of asymmetry in the distribution.

### [Problem 6] Confirming distribution

Use seaborn's sns.displot() and sns.histplot() to display the distribution of objective variables and calculate "kurtosis" and "skewness".

```python
target = df["SalePrice"]
print("skewness= {:.2f}  sharpness = {:.2f}".format(target.skew(), target.kurtosis()))
display(sns.distplot(target, kde = False , rug=False, bins=10))
```

Perform a logarithmic transformation on the objective variable.

```python
target_log = target.apply(np.log)
print("skewness= {:.2f}  sharpness = {:.2f}".format(target_log.skew(), target_log.kurtosis()))
display(sns.distplot(target_log, kde = False , rug = False, bins=10))
```

The original data had a kurtosis of 6.53, but the logarithmic transformation resulted in a kurtosis of 0.81The distribution became closer to a normal distribution

### [Problem 7] Confirming the correlation coefficient

Create a heat map of the correlation coefficient matrix

```python
df_corr = df.corr()
colormap = plt.cm.RdBu
plt.subplots(figsize =(15, 12))
plt.title('Correlation matrix')
sns.heatmap(df_corr, linewidths =0.1,vmax=1.0, square = True, cmap = colormap, linecolor = 'white', annot = False)
plt.show()
```

From the above Heatmap, Variables with high correltion are have bright to dark blue color
